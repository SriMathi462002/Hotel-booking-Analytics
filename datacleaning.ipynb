```python
import pandas as pd
a=pd.read_csv('E:\\hotel_bookingsnew.csv')
```

```python
a.dropna(how="all", inplace=True)
a.shape
```

```python
#a.drop(columns=['arrival_date_year','arrival_date_month','arrival_date_week_number','arrival_date_day_of_month'],axis=1,inplace=True)
```

```python
a['day']=a['day'].ffill()
```

```python
a.shape
```

```python
mean=a.agent.mean()
print(mean)
a.agent.fillna(mean,inplace=True)
```

```python
mode=a.company.mode()
print(mode)
a.company.fillna(mode,inplace=True)
```

```python
a.customer_type.fillna('unknown',inplace=True)
```

```python
a['day']=pd.to_datetime(a['day'], format='%d-%m-%Y')
```

```python
a['Stay_duration']=a.stays_in_weekend_nights+a.stays_in_week_nights
```

```python
a.is_canceled.fillna(0,inplace=True)
a['Booking_Status']=a['is_canceled'].astype(int).map({0:'Not canceled',1:'Canceled'})
```

```python
a.duplicated()
a.drop_duplicates(inplace=True)
```

```python
a.duplicated()
a.head()
```

```python
a.to_csv('E:\\hotel_bookingscleaning.csv',header=True,mode='w')
```
