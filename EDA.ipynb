```python
monthly = df.groupby("arrival_date_month")["hotel"].count()

monthly.plot(kind='line',c='m',marker='o',mfc='y',mec='g')
plt.title("Monthly Booking Trend")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.show()
```

Description :

-Analyzed the number of bookings across different months.

Purpose :

Identify seasonal demand patterns

-Detect peak and off-peak booking periods

```python
df["hotel"].value_counts().plot(kind="bar")
plt.title("Bookings by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Bookings")
plt.show()
```

Description :

-Compared total bookings between City Hotels and Resort Hotels.

Purpose :

-Understand which hotel type attracts more customers

-Helps in strategic planning and resource allocation

```python
cancel_rate = df.groupby("hotel")["is_canceled"].mean()
cancel_rate.plot(kind="bar")
plt.title("Cancellation Rate by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Cancellation Rate")
plt.show()
```

Description :

-Calculated cancellation rate separately for each hotel type.

Purpose :

-Identify hotel types with higher cancellation risk

-Supports policy and pricing decisions

```python
df["market_segment"].value_counts().plot(kind="bar")
plt.title("Bookings by Market Segment",c='r')
plt.xlabel("Market Segment")
plt.ylabel("Bookings")
plt.xticks(rotation=45)
plt.show()
```

Description :

-Analyzed bookings across different market segments such as Online TA, Direct,
and Offline TA/TO.

Purpose :

-Identify high-performing customer acquisition channels

-Helps focus marketing and partnerships

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries by Bookings")
plt.xlabel("Country")
plt.ylabel("Bookings")
plt.show()
```

Description :

-Identified top 10 countries contributing the highest number of bookings.

Purpose :

-Understand geographic demand concentration

-Enables targeted marketing by region

```python
df["adr"].hist(bins=20)
plt.title("ADR Distribution")
plt.xlabel("Average Daily Rate")
plt.ylabel("Frequency")
plt.show()
```

Description :

-Analyzed the distribution of average daily room prices.

Purpose :

-Understand pricing patterns

-Identify common price ranges and potential outliers

```python
df["Stay_duration"].hist(bins=20)
plt.title("Total Stay Duration")
plt.xlabel("Number of Nights")
plt.ylabel("Frequency")
plt.show()
```
